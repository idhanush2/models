```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Amazon_Sale_Report.csv', low_memory = False)

df.head(15)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop('Unnamed: 22', axis = 1, inplace = True)
df.drop('fulfilled-by', axis = 1, inplace = True)
df.drop('B2B', axis = 1, inplace = True)
df.drop('promotion-ids', axis = 1, inplace = True)
df.drop('ship-country', axis = 1, inplace = True)
df.drop('currency', axis = 1, inplace = True)
df.drop('SKU', axis = 1, inplace = True)
df.drop('ship-service-level', axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.drop('ASIN', axis = 1, inplace = True)
df.drop('Fulfilment', axis = 1, inplace = True)
```

```python
df.head(10)
```

```python
df.set_index('index', inplace = True)

df.head()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df['Order ID'] = df['Order ID'].astype(str).str.replace('-', '').astype(int)
df['ship-postal-code'] = df['ship-postal-code'].astype(int)
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype('datetime64[ns]')

df['Day'] = df['Date'].dt.day.astype(int)
df['Month'] = df['Date'].dt.month.astype(int)
df['Year'] = df['Date'].dt.year.astype(int)

```

```python
df.drop('Date', axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.drop('Style', axis = 1, inplace = True)
df.drop('ship-postal-code', axis = 1, inplace = True)
df.drop('Order ID', axis = 1, inplace = True)
df.drop('Courier Status', axis = 1, inplace = True)
```

```python
df.drop('Status', axis = 1, inplace = True)
```
