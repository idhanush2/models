```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    accuracy_score,
    confusion_matrix,
    classification_report,
    r2_score
)
```

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)  

print("MAE:", mae)
print("RMSE:", rmse)
print("R-squared:", r2) 
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X_train_fast = X_train[:3000]
y_train_fast = y_train.iloc[:3000]
#only for faster svms
```

```python
svc_linear = SVC(kernel="linear", C=0.5)
svc_linear.fit(X_train, y_train)

y_pred_linear = svc_linear.predict(X_test)

print("Linear Accuracy:", accuracy_score(y_test, y_pred_linear))
#kernel = linear
print(confusion_matrix(y_test, y_pred_linear))
print(classification_report(y_test, y_pred_linear))
```

```python
svc_rbf = SVC(
    kernel="rbf",
    C=0.5,
    gamma="scale"
)

svc_rbf.fit(X_train_fast, y_train_fast)

y_pred_rbf = svc_rbf.predict(X_test)

print("RBF Accuracy:", accuracy_score(y_test, y_pred_rbf))
#kernel = rbf
print(confusion_matrix(y_test, y_pred_rbf))
print(classification_report(y_test, y_pred_rbf))
```

```python
svc_poly = SVC(
    kernel="poly",
    degree=2,
    C=0.5,
    gamma="scale"
)

svc_poly.fit(X_train_fast, y_train_fast)

y_pred_poly = svc_poly.predict(X_test)

print("Polynomial Accuracy:", accuracy_score(y_test, y_pred_poly))
#kernel = polynomial
print(confusion_matrix(y_test, y_pred_poly))
print(classification_report(y_test, y_pred_poly))
```

```python
from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression()
logistic.fit(X_train, y_train)
y_pred_log = logistic.predict(X_test)
print(y_pred_log)
```

```python
score = accuracy_score(y_pred_log, y_test)
print("Accuracy score: ", score)
print("Classification Report: \n", classification_report(y_pred_log, y_test))
print("Confusion matrix: \n", confusion_matrix(y_pred_log, y_test))
```

```python

model = LogisticRegression()

penalty = ['l1', 'l2', 'elasticnet'] #regularization : lasso, ridge and elastic net

c_values=[100, 10, 1.0, 0.1, 0.01] # inverse of lambda values

solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params = dict(penalty=penalty, C=c_values, solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator = model, param_grid = params, scoring = 'accuracy', cv = 5)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred_gs = grid.predict(X_test)

score = accuracy_score(y_pred_gs, y_test)
print(score)
print(classification_report(y_pred_gs, y_test))
print(confusion_matrix(y_pred_gs, y_test))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred_gs); 
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
clf = DecisionTreeClassifier(random_state = 55) 

clf.fit(X_train, y_train)
```

```python
y_pred_dt = clf.predict(X_test) 

accuracy = accuracy_score(y_test, y_pred_dt) 
print(f"Accuracy: {accuracy :2f}")

report = classification_report(y_test, y_pred_dt) 
print("Classification Report : \n", report)
```

```python
param_grid = {   
    'criterion' : ['gini', 'entropy'],   
    'max_depth' : [None, 5, 10, 15], 
    'min_samples_split' : [2, 5, 10], 
    'min_samples_leaf' : [1, 2, 4] 
}

grid_search = GridSearchCV(estimator = clf, param_grid = param_grid, cv = 5) #initializing gridsearchcv

grid_search.fit(X_train, y_train) 

best_params = grid_search.best_params_   
print("Best Hyperparameters:", best_params)
```

```python
best_clf = DecisionTreeClassifier(**best_params, random_state = 55) 

best_clf.fit(X_train, y_train) 

y_pred_optimized = best_clf.predict(X_test) 

accuracy_optimized = accuracy_score(y_test, y_pred_optimized) 
print(f"Accuracy (Optimized): {accuracy_optimized:.5f}")

report = classification_report(y_test, y_pred_optimized) 
print("Classification Report:\n", report)
```

```python
#using grid search cv
y_pred_optimized = grid_search.predict(X_test) 

accuracy_optimized = accuracy_score(y_test, y_pred_optimized) 
print(f"Accuracy (Optimized) : {accuracy_optimized:.5f}")

report = classification_report(y_test, y_pred_optimized) 
print("Classification Report:\n", report)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_optimized); 
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=55)

rf.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
```

```python
best_rf = grid_search.best_estimator_

y_pred_best = best_rf.predict(X_test)

print("Random Forest Accuracy (With Tuning):", accuracy_score(y_test, y_pred_best))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_best))
print("Classification Report:\n", classification_report(y_test, y_pred_best))
```
